```python
import numpy as np
import matplotlib.pyplot as plt
import circle_fit as cf

center = np.array([0, 0]) # x, y
radius = 1

numberPoints= 100

timePoints = np.linspace(0, 2 * np.pi, numberPoints)

xCircle = center[0] + radius * np.cos(timePoints)
yCircle = center[1] + radius * np.sin(timePoints)

noiseScale = 0.05

xCircelWithNoise = center[0] + (np.random.rand(len(xCircle)) * noiseScale * radius + radius) * np.cos(timePoints)
yCircelWithNoise = center[1] + (np.random.rand(len(xCircle)) * noiseScale * radius + radius) * np.sin(timePoints)


dataArray = np.concatenate((np.array([xCircelWithNoise]), np.array([yCircelWithNoise]))).transpose()


xc, yc, r, _ = cf.least_squares_circle(dataArray)

xCircleEstimate = xc + r * np.cos(timePoints)
yCircleEstimate = yc + r * np.sin(timePoints)

plt.plot(xCircle, yCircle)
plt.plot(xCircelWithNoise, yCircelWithNoise)
plt.plot(xCircleEstimate, yCircleEstimate)

plt.legend(["Orig", "WithNoise", "Estimate"])
```
